```python
import pandas as pd
data = pd.read_csv('/content/FIFA World Cup Attendance.csv')
data
```

```python
del data['Unnamed: 0']
```

```python
data
```

```python
# which country has hosted how many times
data1 = data
data1 = data1.drop(22)
data1.Hosts.value_counts()
```

```python
## top 10 hosts according to count on chart

hosts = data1.Hosts.value_counts()
TOP_10_HOSTS_ACCORDING_TO_COUNT = hosts.head(10).plot(autopct='%1.0f%%', kind='pie',figsize=(24,5),title="TOP 10 HOSTS")
```

```python
### TOP 10 Total_Attendance in any year
data2 = data1
total_attendance=data2.groupby(['Year','Hosts'])['Total_Attendance'].max().reset_index().sort_values('Total_Attendance',ascending=False).head(10)
total_attendance
```

```python
#Total attendance in millions
total_attendance1 =data2.groupby(['Year'])['Total_Attendance'].max().reset_index().sort_values('Year',ascending=True)
total_attendance1.plot(x='Year',y="Total_Attendance",figsize=(12,5),title="Total Attendance in Millions over the years")
```

```python
#TOP 10 least Number of matches played in any year
data3 = data1
number_of_matches =data3.groupby(['Year'])['Matches'].min().reset_index().sort_values('Matches',ascending=True).head(10)
number_of_matches
```

```python
#Total number of matches
c = ['red','orange','yellow','green','black','blue']
number_of_matches1 =data3.groupby(['Year'])['Matches'].max().reset_index().sort_values('Year',ascending=True)
number_of_matches1.plot(x='Year',y="Matches",figsize=(12,5),title="Total Number of Matches Played over the years",kind="bar",color=c)
```

```python
#Top 10 Average Attendance in any year
data4= data1
avg_number_of_attendance1 =data4.groupby(['Year','Hosts'])['Average_Attendance'].max().reset_index().sort_values('Average_Attendance',ascending=False).head(10)
avg_number_of_attendance1 
```

```python
#Average attendance
avg_number_of_attendance2 =data4.groupby(['Year'])['Average_Attendance'].max().reset_index().sort_values('Year',ascending=True)
avg_number_of_attendance2.plot(x='Year',y="Average_Attendance",figsize=(12,5),title="Average number of Attendance over the years")
```
